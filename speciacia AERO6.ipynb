```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read gspro file https://www.airqualitymodeling.org/index.php/CMAQv5.0_GSPRO_Example
speciacia=pd.read_csv('/home/p6001/fume_project/example_data/input/static_data/speciations/gspro_AE6_pokus2.csv', sep=';')
```

```python
#read gsref file https://www.airqualitymodeling.org/index.php/CMAQv5.0_GSREF_example
kody=pd.read_csv('/home/p6001/fume_project/example_data/input/static_data/speciations/gsref', sep=';',header=-1)
```

```python
for i in speciacia.cislo.unique():
    ax=speciacia[speciacia['cislo']==i].plot.bar(x='spec',y='fac',legend=False,title='{0}--{1}'.format(i,str(kody[kody[1]==i][9].head(1))[-56:-23]), fontsize=18, figsize=(10, 5))
    ax.set_xlabel("Species", fontsize=18)
    plt.show()
    plt.close()
```
