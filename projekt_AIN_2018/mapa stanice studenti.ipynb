```python
import pandas as pd
import folium
import geopandas as gpd
import shapely
```

```python
nove=gpd.read_file('/media/sf_Share/staniceshp/stanice_nove.shp', encoding='utf-8')
lat_nove = nove['Y'][:]
lon_nove = nove['X'][:]
typ_nove = nove['Typ stanic'][:]
name_nove=nove['Umiestneni'][:]

vsetky=gpd.read_file('/media/sf_Share/staniceshp/stanice_projekt.shp', encoding='utf-8')

lat_vsetky = vsetky['Y'][:]
lon_vsetky = vsetky['X'][:]
typ_vsetky = vsetky['Typ stanic'][:]
name_vsetky=vsetky['Umiestneni'][:]
```

```python
map = folium.Map(location=[48.85534, 19.7367], zoom_start=8, tiles="Stamen Terrain")

    
fgvsetky = folium.FeatureGroup(name="Všetky stanice")

for lt, ln, typ, name_st  in zip(lat_vsetky, lon_vsetky, typ_vsetky, name_vsetky):
    fgvsetky.add_child(folium.CircleMarker(location=[lt, ln], radius = 6, popup=name_st+' typ stanice: '+str(typ),
    fill_color='yellow', fill=True,  color = 'black', fill_opacity=0.7))     
    
  
    
    
    
fgnove = folium.FeatureGroup(name="Planované stanice")

for lt, ln, typ, name_st  in zip(lat_nove, lon_nove, typ_nove, name_nove):
    fgnove.add_child(folium.CircleMarker(location=[lt, ln], radius = 6, popup=name_st+', typ stanice: '+str(typ),
    fill_color='red', fill=True,  color = 'black', fill_opacity=0.7))    
   
   
map.add_child(fgvsetky)
map.add_child(fgnove)
map.add_child(folium.LayerControl())

#map.save("Mapa2T.html")
display(map)
```
